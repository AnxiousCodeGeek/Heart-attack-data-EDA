```python
import plotly.offline as py
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import plotly.express as px

def attack_ratio(data):
    """
    Make a pie chart of 'Result' values
    Results: 0 = Negative, 1 = Positive
    """
    Result = data['Result'].value_counts()
    values = [Result[0], Result[1]]
    labels = ['Negative', 'Positive']
    colors = ['3A0CA3', 'ba181b']
    
    fig_pie = go.Pie(labels=labels, values=values,
                     marker={'colors': colors, 
                             'line': {'color': 'White', 'width': 2}})
    layout = go.Layout(height=500, width=500, title={'text': 'Heart Attack Ratio Proportion'})  # Set the height and width
    fig = go.Figure(data=[fig_pie], layout=layout)
    py.iplot(fig)
    return py
    
    
attack_ratio(heartattackeda)
```

Most of the cases diagnosed are likely to have a heart attack , more than 50 %. While less than 40 % cases are negative.

```python


def gender_ratio(data):
    """
    Make a pie chart of 'Gender' values
    Gender: 0 = Female, 1 = Male
    """
    # Pie chart of Gender ratio
    Result = data['Gender'].value_counts()
    total_count = Result.sum()
    values = {'Female': Result[0] / total_count, 'Male': Result[1] / total_count}
    
    fig_pie = px.pie(values=values,
                     names=values.keys(),
                     height=500, width=500, 
                     hole=0.7,
                     color_discrete_sequence = ['#4c78a8', '#72b7b2']
                     )
    fig_pie.update_layout(plot_bgcolor='#fafafa', paper_bgcolor='#fafafa',
                            title={'text': 'Ratio of Heart Attack in Males and Females'})   
    
    return fig_pie


fig_pie = gender_ratio(heartattackeda)
fig_pie.show()
```

The number of Male population are two times of Female population.

```python
# Let's calculate the proportion of Positive heart attack cases for both Female and Male.

def attack_proportion_gender(data):
    """
    Plot a bar chart of the proportion of Result vs. Gender.
    Show the percentage of positive heart attack Result for each gender.
    """
    # Bivariate analysis: Gender vs. Result          
    data['Gender'].groupby(data['Result']).value_counts(normalize=True).rename('proportion').reset_index().pipe((sns.barplot, 'data'), 
                                                                                                                 x='Gender', y='proportion', hue='Result', palette='Set1')
    plt.title('Proportion of Heart Attack Result for Gender')
    plt.xlabel('Gender (0 = Negative, 1 = Positive)')
    plt.show()

    
attack_proportion_gender(heartattackeda)
```

```python
# Show distributions of risk factors, Systolic blood pressure, Diastolic blood pressure, Heart rate and Troponin

def risk_factors_dist(data):
  
    fig = plt.figure(figsize=(18, 8))
    
    # Resting blood pressure
    plt.subplot(5, 2, 1)
    sns.distplot(data['Systolic blood pressure'])
    plt.title('Systolic blood pressure (mmHg) Distribution', fontsize=8)
    plt.axvline(x=130, color='r', linestyle='--', label='Isolated Systolic Hypertension: over 130 mmHg')
    plt.ylabel('')
    plt.legend()
    
    plt.subplot(5, 2, 2)
    sns.boxplot(data['Systolic blood pressure'], orient='h')   
        
        
    # Serum Diastolic Blood Pressure 
    plt.subplot(5, 2, 3)
    sns.distplot(data['Diastolic blood pressure'])
    plt.title('Diastolic blood pressure (mmHg) Distribution', fontsize=8)
    plt.axvline(x=80, color='r', linestyle='--', label='Isolated Diastolic Hypertension : over 80-90 mmHg')
    plt.ylabel('')
    plt.legend()
    
    plt.subplot(5, 2, 4)
    sns.boxplot(data['Diastolic blood pressure'], orient='h')
  
    
    # Maximum heart rate
    plt.subplot(5, 2, 5)
    sns.distplot(data['Heart rate'])
    plt.title('Heart Rate (bpm) Distribution', fontsize=8)
    plt.ylabel('')
    plt.xlim(0, 200)  # Set x-axis limits to 0-200
    
    plt.subplot(5, 2, 6)
    sns.boxplot(data['Heart rate'], orient='h')
    plt.xlim(0, 200)  # Set x-axis limits to 0-200
    
    plt.subplot(5, 2, 7)
    sns.distplot(data['Troponin'])
    plt.title('Troponin (ng/ml) Distribution', fontsize=8)
    plt.axvline(x=0.40, color='r', linestyle='--', label='Indication of myocardial infarction : over 0.40 (ng/ml)')
    plt.xlim(-2,4)
    plt.ylabel('')
    plt.legend()
    
    plt.subplot(5, 2, 8)
    sns.boxplot(data['Troponin'], orient='h')
    plt.xlim(-2,12)
    
    plt.subplot(5, 2, 9)
    sns.distplot(data['CK-MB_normalized'])
    plt.title('Serum CK-MB normalized (U/L) Distribution', fontsize=8)
    plt.axvline(x=30, color='r', linestyle='--', label='Indication of myocardial injury : over 30 (U/L)')
    plt.xlim(-0.5,2)
    plt.ylabel('')
    plt.legend()
    
    plt.subplot(5, 2, 10)
    sns.boxplot(data['CK-MB_normalized'], orient='h')
    # plt.xlim(-2,12)

    plt.tight_layout()
    plt.show()
    
    
risk_factors_dist(heartattackeda)
```

A common analysis question: 
**What is the average age of people who have a heart failure python?**

The average age of people who have heart failure is 53.51 years old. This information is based on the analysis of a dataset containing 12 features that can be used to predict mortality by heart failure. The dataset includes information on age, Gender, chest pain type, resting blood pressure, serum Diastolic Blood Pressure, fasting blood sugar, resting electrocardiogram (ECG) results, maximum heart rate achieved, exercise-induced angina, old peak (ST depression induced by exercise relative to rest), slope of the peak exercise ST segment, and heart disease. The heart disease variable represents the output class, with 1 indicating heart disease and 0 indicating normal. The dataset was analyzed using Python, and the average age of people with heart disease was calculated to be 53.51 years old.

**Feature Variable - Age**

```python
#  Make a standard distribution and distributions for Gender

def age_distribution(data):
   
    fig = plt.figure(figsize=(15, 10))

    # Age distribution
    plt.subplot(2, 4, 1)
    sns.distplot(data['Age']) 
    plt.title('Age Distribution', fontsize=15)
    plt.ylabel('')  # Add this line to remove the y-axis label

    plt.subplot(2, 4, 5)
    sns.boxplot(data['Age'], orient='h')
    

    # Age distribution for Gender
    plt.subplot(2, 4, 2)
    sns.histplot(data=data, x='Age', hue='Gender', element='step', palette=['Red', 'Blue'], kde=True)
    plt.title('Age Distribution (Male vs. Female)', fontsize=15)
    plt.xlim(0,120)
    plt.ylabel('')  # Add this line to remove the y-axis label

    plt.subplot(2, 4, 6)
    sns.boxplot(x=data['Age'], y=data['Gender'], 
                palette='Set1', orient='h', showfliers=False)
    
    
    # Age distribution for Result
    plt.subplot(2, 4, 3)
    Negative = heartattackeda[heartattackeda['Result']==0]['Age']
    Positive = heartattackeda[heartattackeda['Result']==1]['Age']

    sns.distplot(Negative, color='Green', label='Negative')
    sns.distplot(Positive, color='Red', label='Positive')
    plt.title('Age Distribution for Result', fontsize=15)
    plt.legend(title='Result', fontsize=10)
    plt.ylabel('')  # Add this line to remove the y-axis label
    
    plt.subplot(2, 4, 7)
    sns.boxplot(x=data['Age'], y=data['Result'], orient='h', palette='Dark2')
    
    # Age distribution for troponin
    plt.subplot(2, 4, 4)
    sns.histplot(data=data, x='Age', hue='Troponin', element='step', palette=['Red', 'Blue'], kde=True)
    plt.title('Troponin vs Age')
    plt.gca().legend(title='Troponin', labels=['Negative','Positive'])
    plt.ylabel('')

    # plt.subplot(2, 4, 8)
    # # plt.xlim(-1,10)
    # sns.boxplot(x = data['Age'], y=data['Troponin'],
    #             orient='h', palette='Dark2')
    
    # Not plotting boxplot for troponin due to complications
    # Dropped CK_MB normalized as the attributes doesn't seem to contribute much to the heart attack ratio
    
    plt.tight_layout()
    plt.show()

    
age_distribution(heartattackeda)
```

```python
# Effects of features for heart attack by age

def numeric_features_vs_age(data): 
    Negative = data.Result==0
    Positive = data.Result==1
    
    age_Negative = data.Age[Negative]
    age_Positive = data.Age[Positive]
    
    b_color = 'DarkBlue'
    m_color = 'Green'
    
    fig = plt.figure(figsize=(12, 10))
    
    # Systolic blood pressure (mmHg) 
    plt.subplot(2, 2, 1)
    plt.scatter(x=age_Negative, y=data['Systolic blood pressure'][Negative], color=b_color)
    plt.scatter(x=age_Positive, y=data['Systolic blood pressure'][Positive], color=m_color)
    plt.title('Systolic blood pressure vs. age', fontsize=15)
    plt.legend(['Negative', 'Positive'])
    plt.xlabel('age', fontsize=10)
    plt.ylabel('Systolic blood pressure (mmHg)', fontsize=10)
        
        
    # Diastolic blood pressure (mg/dl) 
    plt.subplot(2, 2, 2)
    plt.scatter(x=age_Negative, y=data['Diastolic blood pressure'][Negative], color=b_color)
    plt.scatter(x=age_Positive, y=data['Diastolic blood pressure'][Positive], color=m_color)
    plt.title('Diastolic blood pressure (mmHg)', fontsize=15)
    plt.legend(['Negative', 'Positive'])
    plt.xlabel('age', fontsize=10)
    plt.ylabel('Diastolic Blood Pressure (mmHg)', fontsize=10)
 

    # Heart rate achieved (bpm)
    plt.subplot(2, 2, 3)
    plt.scatter(x=age_Negative, y=data['Heart rate'][Negative], color=b_color)
    plt.scatter(x=age_Positive, y=data['Heart rate'][Positive], color=m_color)
    plt.title('Heart rate vs. age', fontsize=15)
    plt.legend(['Negative', 'Positive'])
    plt.xlabel('age',fontsize=10)
    plt.ylabel('Maximum heart rate achieved (bpm)', fontsize=10)
    plt.ylim(0,200)
    
    # troponin
    plt.subplot(2, 2, 4)
    plt.scatter(x=age_Negative, y=data['Troponin'][Negative], color = b_color)
    plt.scatter(x=age_Positive, y=data['Troponin'][Positive], color = m_color)
    plt.title('Troponin vs. age', fontsize=15)
    plt.legend(['Negative', 'Positive'])
    plt.xlabel('age',fontsize=10)
    plt.ylabel('Troponin', fontsize=10)
    
    plt.tight_layout()
    plt.show()


numeric_features_vs_age(heartattackeda)
```

```python
import numpy as np

def correlation_heatmap(data):
    """
    Correlation Heat Map of the features.
    """
    numerical_data = data.select_dtypes(include=[np.number])
    plt.figure(figsize=(11, 7))
    sns.heatmap(numerical_data.corr(), annot=True, linewidth=0.2, 
                fmt='.2f', cmap='RdGy_r')
    plt.title('Correlations between Features', fontsize=15)
    plt.show()
    
    
correlation_heatmap(heartattackeda)
```
